# Plot KL divergence for solvers

```python
import torch
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Hyper-Parameters

```python
EPS=1 # 0.1, 1
DIM=16 # 16

ylims = [-3,70] if EPS==0.1 else [-3,100]
```

```python
joint_forward = torch.load(
           f"./data_for_plots/eps_{EPS}/KL/dim_{DIM}/kl_joint_forward.pth")

joint_rev = torch.load(
           f"./data_for_plots/eps_{EPS}/KL/dim_{DIM}/kl_joint_reverse.pth")
 
enot_forward = torch.load( 
           f"./data_for_plots/eps_{EPS}/KL/dim_{DIM}/kl_enot_forward.pth")
enot_rev = torch.load(
           f"./data_for_plots/eps_{EPS}/KL/dim_{DIM}/kl_enot_reverse.pth")

vargas_forward = torch.load( 
           f"./data_for_plots/eps_{EPS}/KL/dim_{DIM}/kl_vargas_forward.pth")
vargas_rev = torch.load(
           f"./data_for_plots/eps_{EPS}/KL/dim_{DIM}/kl_vargas_reverse.pth")

if EPS==0.1:
    alter_forward = torch.load( 
               f"./data_for_plots/eps_{EPS}/KL/dim_{DIM}/kl_alter_forward.pth")
    alter_rev = torch.load(
               f"./data_for_plots/eps_{EPS}/KL/dim_{DIM}/kl_alter_reverse.pth")

    bortoli_forward = torch.load( 
               f"./data_for_plots/eps_{EPS}/KL/dim_{DIM}/kl_bortoli_forward.pth")

    bortoli_rev =  torch.load(
               f"./data_for_plots/eps_{EPS}/KL/dim_{DIM}/kl_bortoli_reverse.pth")
    
    
   
 
```

## KL forward plot

```python
fig,ax = plt.subplots(1,1,figsize=(5,3),dpi=200)
if EPS==0.1:
    ax.plot(alter_forward,color='black',label=r'$\lfloor$'+ 'FB-SDE-A' + r'$\rceil$')
    ax.plot(bortoli_forward,color='darkorange',label=r'$\lfloor$'+ 'DiffSB' + r'$\rceil$' )
ax.plot(joint_forward,color='blue',label=r'$\lfloor$'+ 'FB-SDE-J' + r'$\rceil$' )
ax.plot(enot_forward,color='darkgreen',label=r'$\lfloor$'+ 'ENOT' + r'$\rceil$' )
ax.plot(vargas_forward,color='firebrick',label=r'$\lfloor$'+ 'MLE-SB' + r'$\rceil$' )
ax.set_xlabel('t, time')
ax.set_ylabel(r"$\frac{1}{2\epsilon}\mathcal{L}^{2}_{fwd}[t]$",rotation='vertical',fontsize=15, loc='center');
ax.set_ylim(ylims)
legend_properties = {'weight':'bold'}
ax.legend(prop=legend_properties,loc='upper left')
ax.grid();
fig.tight_layout(pad=0.5)
```

## KL Reverse plot

```python
fig,ax = plt.subplots(1,1,figsize=(5,3),dpi=200)
if EPS==0.1:
    ax.plot(alter_rev,color='black',label=r'$\lfloor$'+ 'FB-SDE-A' + r'$\rceil$')
    ax.plot(bortoli_rev,color='darkorange',label=r'$\lfloor$'+ 'DiffSB' + r'$\rceil$' )
ax.plot(joint_rev,color='blue',label=r'$\lfloor$'+ 'FB-SDE-J' + r'$\rceil$' )
ax.plot(enot_rev,color='darkgreen',label=r'$\lfloor$'+ 'ENOT' + r'$\rceil$' )
ax.plot(vargas_rev,color='firebrick',label=r'$\lfloor$'+ 'MLE-SB' + r'$\rceil$' )
ax.set_xlabel('t, time')
ax.set_ylabel(r"$\frac{1}{2\epsilon}\mathcal{L}^{2}_{rev}[t]$",rotation='vertical',fontsize=15, loc='center');
ax.set_ylim(ylims)
legend_properties = {'weight':'bold'}
ax.legend(prop=legend_properties,loc='upper left')
ax.grid();
fig.tight_layout(pad=0.5)
```
